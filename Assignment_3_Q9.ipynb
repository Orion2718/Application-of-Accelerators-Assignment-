```python
# Question 9

import numpy as n
import matplotlib.pyplot as pt

a = n.loadtxt("data.txt", dtype=float) # Saving the data in an array

pt.title('Assignment 3 - Q9 Graph') # Title for the graph
pt.grid(color='black', linestyle = '-.', linewidth = 1.7) # For the grid
pt.plot(a,'blue') # Plot Graph
pt.scatter(n.argmax(a), n.max(a), color='red') # Highest point
pt.show() # Display the graph
```
