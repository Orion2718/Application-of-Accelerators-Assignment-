```python
import pandas
import numpy as np
import matplotlib.pyplot as plt


csvFile = pandas.read_csv('2016-07-11_ipm_data.txt')
print(csvFile)

maxi = np.max(csvFile)
print(maxi)

indmx = np.argmax(csvFile)
print(indmx)

plt.plot(csvFile,'blue')
plt.scatter(indmx,maxi,color='red')
plt.title('Graph')
plt.grid(color='black')
plt.show()
```
